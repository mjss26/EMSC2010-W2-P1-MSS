This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
import pandas as pd #used to read and clean data
import matplotlib.pyplot as plt #used for data plotting
```

```python
spreadsheet_name = 'CO2.csv' # Data sourced from Mauna Loa CO2 datasets
df = pd.read_csv(spreadsheet_name) #read the data into a dataframe
```

```python
df.head() #show the dataframe header information
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['year'],df['average'],'k') #plot Year vs. CO2 on ax as a black ('k') line
ax.set_xlabel('Year') #text for the x-axis label
ax.set_ylabel('CO2 (ppm)') #text for the y-axis label
ax.set_xlim([1957,2026]) #interval for the x-axis
ax.set_ylim([310,440]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['decimal date'],df['average'],'k') #plot Year vs. CO2 on ax as a black ('k') line
ax.set_xlabel('Year - Monthly data points') #text for the x-axis label
ax.set_ylabel('CO2 (ppm)') #text for the y-axis label
ax.set_xlim([1957,2026]) #interval for the x-axis
ax.set_ylim([310,440]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.scatter(df['decimal date'], df['average'], c='r', s=10) #plot decimal date vs. CO2 on ax as red ('r') scatter points with size 10
ax.set_xlabel('Decimal Date (Monthly data points)') #text for the x-axis label
ax.set_ylabel('CO2 (ppm)') #text for the y-axis label
ax.set_title('Mauna Loa Monthly CO2 Data 1958-2026') # Add a title to the plot
ax.set_xlim([1957,2026]) #interval for the x-axis
ax.set_ylim([310,440]) #interval for the y-axis
ax.minorticks_on() #show the minor ticks
```

In an attempt to view the individual monthly scatter points I used code from week 3 to create 2 stacked plots - and adjusted scale for both plots


```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, figsize=(20, 6)) # Create a figure with two subplots side-by-side

# --- First Plot: 1958-1996 ---

df_1958_1996 = df[(df['decimal date'] >= 1958) & (df['decimal date'] <= 1996)]
ax[0].scatter(df_1958_1996['decimal date'], df_1958_1996['average'], c='r', s=10)
ax[0].set_xlabel('Decimal Date (Monthly data points)')
ax[0].set_ylabel('CO2 (ppm)')
ax[0].set_title('Mauna Loa Monthly CO2 Data 1958-1996')
ax[0].set_xlim([1957, 1997])
ax[0].set_ylim([300, 400]) # Maintain consistent y-axis for comparison
ax[0].minorticks_on()

# --- Second Plot: 1997-2026 ---

df_1997_2026 = df[(df['decimal date'] >= 1997) & (df['decimal date'] <= 2026)]
ax[1].scatter(df_1997_2026['decimal date'], df_1997_2026['average'], c='r', s=10)
ax[1].set_xlabel('Decimal Date (Monthly data points)')
ax[1].set_ylabel('CO2 (ppm)')
ax[1].set_title('Mauna Loa Monthly CO2 Data 1997-2026')
ax[1].set_xlim([1996, 2026])
ax[1].set_ylim([350, 440]) # Maintain consistent y-axis for comparison
ax[1].minorticks_on()

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12, 6)) # Create a single figure for the plot

ax.plot(df['decimal date'], df['average'], color='blue', linewidth=1) # Plot CO2 data as a line

ax.set_xlabel('Decimal Date (Monthly Data)') # Label the x-axis
ax.set_ylabel('CO2 (ppm)') # Label the y-axis
ax.set_title('Mauna Loa Monthly CO2 Data Over Time (Line Plot)') # Add a title
ax.grid(True, linestyle='--', alpha=0.7) # Add a grid for readability
ax.minorticks_on() # Show minor ticks

plt.tight_layout() # Adjust layout to prevent overlapping elements
plt.show()
```
